# <center>`Linear Regression using Normal Equation`</center>

<b>Normal Equation representation is given by:</b>   
    θ = (X<sup>T</sup> X)<sup>-1</sup> X<sup>T</sup> y    
    where,  
    θ is the values of the parameter and can be suplimented as β in the above mentioned Regression Formulae.   
    X is a matrix nxm that contains values x<sub>1</sub>, x<sub>2</sub>, x<sub>3</sub>, ...  
    X<sup>T</sup> is the transpose of matrix X  
    <sup>-1</sup> represents the inverse of the matrix  
    y denotes the target

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# generate random numbers

np.random.seed(50)
X = 2 * np.random.rand(100, 1)
y = 4 + 3*X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, 'bo')

plt.xlabel("X")
plt.ylabel("y")
plt.axis([0, 2, 0, 15])

plt.show()
```

```python
X.shape
```

Since, our X matrix has only one column (feature/parameter), when using normal equation, we maynot get the values of the intercept. So, we need to add one other column (feature) as a bias term.

```python
# add x = 1 for each instance
Xbias = np.c_[np.ones((100, 1)), X] # concatinates two matrices

# looking at top 5 rows from the total 100 rows
Xbias[:5]
```

```python
# Using Normal Equation
theta = np.linalg.inv(Xbias.T.dot(Xbias)).dot(Xbias.T).dot(y)
```

```python
# gives the value of intercept and coefficient
theta
```

```python
Xnew = np.array([[0], [3]])
XnewBias = np.c_[np.ones((2, 1)), Xnew] # x = 1

y_predict = XnewBias.dot(theta)
y_predict
```

```python
plt.plot(Xnew, y_predict, 'r-', label="Regression Line")
plt.plot(X, y, 'bo')

plt.xlabel("X")
plt.ylabel("y")
plt.axis([0, 2, 0, 12])
plt.legend()

plt.show()
```

Hence, the regression line with minimum error is created.
